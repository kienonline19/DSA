# Lists

Some basic things you can do with lists. We can create a new list and assign it to variable:

```python
var = [14, 34, 89, 10]
```

We can access and change the $k$-th element in the list. Note that elements in the list are indexed from zero:

```python
var[1]
```

```python
var[2] = 45
var
```

`len` function outputs the length of the list:

```python
len(var)
```

We can use append to add new element at the end of the list:

```python
var.append(6)
var
```

Convenient syntax to initialize a list with the same value repated a given number of times:

```python
var = [False] * 30
print(var)
```

You can also multiply any other list by an integer, wich will output list that repeats its content given number of times. This is less useful:

```python
[1,2,7]*5
```

Write code that outputs a product of all elements of the list `var`

```python
# result = product of
# all elements of var
result = 1
for x in var:
    result= result * x
# %%
result
```

Construct a list with elements `[1,2,3,4,..., n]`

```python
n = 25
# produce a list with elements
# [1,2, ..., n]
var = [0] * n
for i in range(1, n+1):
    var[i-1] = i
print(var)
```

Alternative soluttion

```python
var = []
# %%
for i in range(1, n+1):
    var.append(i)
print(var)
```

Check if the list `var` is increasing:

```python
var = [1,2,5, 21, 17]
## check if the list 
# is increasing
var[0]
# %%
result = True
for i in range(len(var) - 1):
    if var[i] >= var[i+1]:
        result = False
        break
print(result)



```

Find the first occurence of element `x` in the list `var`. If there is no such element, output $-1$.

```python
# %%
var = [1,2,4,6,5, 11, 3]
x = 7
res = -1
for i in range(len(var)):
    if var[i] == x:
        res = i
        break
print(res)
```

Check if two lists have element in common

Solution 1, reusing (by copy-pasting) the previous code:

```python
result = False
for x in var1:
    res = -1
    for i in range(len(var2)):
        if var2[i] == x:
            res = i
            break
    if res != -1:
        result = True
```

Solution 2, a bit more compact:

```python
result = False
for x in var1:
    for y in var2:
        if x==y:
            result = True
            # Note that this break exists only the inner loop ! (over y)
            break 
print(result)
```
