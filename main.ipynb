```python
def generate_blocks(n: int) -> list[str]:
    res: list[str] = []
    path: list[str] = []

    def backtrack(open_used: int, close_used: int, consec_open: int) -> None:
        if open_used == n and close_used == n:
            res.append("".join(path))
            return

        # Try '(' first to get lexicographic order
        if open_used < n and consec_open < 2:
            path.append('(')
            backtrack(open_used + 1, close_used, consec_open + 1)
            path.pop()

        # Try ')'
        if close_used < open_used:
            path.append(')')
            backtrack(open_used, close_used + 1, 0)  # reset consecutive '(' count
            path.pop()

    backtrack(0, 0, 0)
    return res

generate_blocks(4)
```
